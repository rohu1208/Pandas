# Merge and Join operation

```python
data1 = pd.DataFrame({'key1':[1,2,4,5,6],
                      'key2':[4,5,6,7,8],
                      'key3':[3,4,5,6,6]
})
```

```python
data2 = pd.DataFrame({'key1':[1,2,45,6,67],
                      'key4':[56,5,6,7,8],
                      'key5':[3,56,5,6,6]
}
)
```

```python
data1
```

```python
data2
```

```python
pd.merge([data1,data2])
```

```python
pd.merge(data1,data2)
```

```python
#merge based on inner node i.e, intersection,and diff operations as right joint by default calculate innermerge 
#key1 is common try to create intersection 
pd.merge(data1,data2,how='left')
```

```python
pd.merge(data1,data2,how='right')
```

```python
pd.merge(data1,data2,how='outer')
```

```python
pd.merge(data1,data2,how='cross')
```

```python
pd.merge(data1,data2,how='left',left_on='key1',right_on='key4')
```

```python
data1
```

```python
data2
```
